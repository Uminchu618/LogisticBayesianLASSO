```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Noto Sans JP"

# 保存したCSVファイル (beta_estimates.csv) を読み込む
df = pd.read_csv("beta_estimates.csv")

# 真のパラメータ
beta_true = [0, 1.0, -2.0, 0.1]

# 各パラメータのヒストグラムをプロット
fig, axes = plt.subplots(1, 4, figsize=(15, 4))
for i, ax in enumerate(axes):
    ax.hist(df.iloc[:, i], bins=20, alpha=0.7)
    ax.axvline(
        beta_true[i], color="red", linestyle="dashed", linewidth=2
    )  # 真値の垂直線
    ax.set_title(f"Beta {i}")
    ax.set_xlabel("推定値")
    ax.set_ylabel("頻度")
fig.suptitle("Sample Distribution of Estimated Parameters lambda=1.0")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Lasso Path のプロット用にCSVファイル (lasso_path_estimates.csv) を読み込む
df_path = pd.read_csv("lasso_path_estimates.csv")

plt.figure(figsize=(8, 6))
for col, label in zip(["beta0", "beta1", "beta2", "beta3"], ["β0", "β1", "β2", "β3"]):
    plt.plot(df_path["lambda1"], df_path[col], marker="o", label=label)
plt.xscale("log")
plt.xlabel("lambda1")
plt.ylabel("回帰係数の平均推定値")
plt.title("Lasso Path")
plt.legend()
plt.grid(True)
plt.show()
```
